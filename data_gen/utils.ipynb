```python
import os
import shutil
import numpy as np
from random import randint

train_dir = "train"
validate_dir = "validate"
test_dir = "test"

train_frac = 0.6
validate_frac = 0.2
test_frac = 0.2
        
def make_dest_dirs(dest_dirs):
    for d in dest_dirs:
        #if os.path.exists(d):
        #    shutil.rmtree(d)
        if not os.path.exists(d):
            os.makedirs(d)
    
def generate_data_dirs(source_dir, dest_dir, label):
    files = os.listdir(source_dir)
    num_files = len(files)
    num_train = int(train_frac * num_files)
    num_validate = int(validate_frac * num_files)
    
    train_dir_path = os.path.join(dest_dir, train_dir)
    validate_dir_path = os.path.join(dest_dir, validate_dir)
    test_dir_path = os.path.join(dest_dir, test_dir)
    
    make_dest_dirs([train_dir_path, validate_dir_path])

    for i, f in enumerate(files):
        # skip bmp file, if such     
        _, ext = os.path.splitext(f)
        if ext == ".bmp":
            continue
        
        full_path = os.path.join(source_dir, f)
        
        print ("copying file no %d %s" % (i, full_path))
        
        if i < num_train:
            dest_subdir = train_dir_path
        elif i >= num_train and i < num_train + num_validate:
            dest_subdir = validate_dir_path
        else:
            dest_subdir = test_dir_path
            
        dest_file_path = os.path.join(dest_subdir, label)
        if not os.path.exists(dest_file_path):
            os.makedirs(dest_file_path)
            
        # add random int to avoid name collision from different batches
        rand_num = randint(0, 1000)
        name, ext = os.path.splitext(f)
        dest_file_name = os.path.join(dest_file_path, "%s_%d%s" % (name, rand_num, ext))
        shutil.copy2(full_path, dest_file_name)

def run_for_single_batch(dir_path, batch_num):
    out_dir_path = os.path.abspath(os.path.join(dir_path, ".."))
    stegged_label = "stegged"
    not_stegged_label = "not_stegged"
    stegged_dir = os.path.join(dir_path, "%s%d" % (stegged_label, batch_num))
    not_stegged_dir = os.path.join(dir_path, "%s%d" % (not_stegged_label, batch_num))

    generate_data_dirs(stegged_dir, out_dir_path, stegged_label)
    generate_data_dirs(not_stegged_dir, out_dir_path, not_stegged_label)
    
```
